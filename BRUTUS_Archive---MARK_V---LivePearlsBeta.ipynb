```python
def brutus2(h,i,j,k,l,m):
    
    
    riskusd = 5.00
    leverage = 100

    symbol_info = mt5.symbol_info(h)
    
    stpips = abs(l - j)*1000
    point = mt5.symbol_info(h).point
    deviation = 20
    symbol = h
    base, quote = symbol[:3], symbol[3:]
    
    threshold = 4
    threshold2 = 1
    iterator = 0
    TTIMER = []
    RATE = []
    RATE2 = []
    
    while iterator <= 30:
        tstart = dt.now()
        if i == "LONG":
            price = mt5.symbol_info_tick(symbol).bid
                if quote == "USD":
                    exchange = 1.00
                else:
                    convertto = quote + "USD"
                    symbol_info = mt5.symbol_info(convertto)
                    if symbol_info is None:
                        convertto2 = "USD" + quote
                        symbol_info = mt5.symbol_info(convertto2)
                        exchange = mt5.symbol_info(convertto2).bid
                    else:
                        exchange = mt5.symbol_info(convertto).bid
                lot = round(riskusd/(stpips*exchange),2)
                request = {
                    "action": mt5.TRADE_ACTION_DEAL,
                    "symbol": symbol,
                    "volume": lot,
                    "type": mt5.ORDER_TYPE_BUY,
                    "price": price,
                    "sl": l,
                    "tp": k,
                    "deviation": deviation,
                    "magic": 234000,
                    "comment": "BRUTUS Mark V",
                    "type_time": mt5.ORDER_TIME_GTC,
                    "type_filling": mt5.ORDER_FILLING_IOC,
                }
            extraj = j + m
            selected=mt5.symbol_select(h,True)
            if not selected:
                print("Failed to select pair")
                mt5.shutdown()
                quit()
                iterator = 30
            lasttickb=mt5.symbol_info_tick(h).bid
            if lasttickb > extraj:
                RATE2.append(lasttickb)
            else:
                RATE2 = []
            if lasttickb > j:
                RATE.append(lasttickb)
            else:
                RATE = []
            print(RATE)
            lrate = len(RATE)
            if lrate >= threshold:
                result = mt5.order_send(request)
                print("1. order_send(): by {} {} lots at {} with deviation={} points".format(symbol,lot,price,deviation));
                iterator = 30
            lrate2 = len(RATE2)
            if lrate2 >= threshold2:
                result = mt5.order_send(request)
                print("1. order_send(): by {} {} lots at {} with deviation={} points".format(symbol,lot,price,deviation));
                iterator = 30
        elif i == "SHORT":
            price = mt5.symbol_info_tick(symbol).ask
                if quote == "USD":
                    exchange = 1.00
                else:
                    convertto = base + "USD"
                    symbol_info = mt5.symbol_info(convertto)
                    if symbol_info is None:
                        convertto2 = "USD" + base
                        symbol_info = mt5.symbol_info(convertto2)
                        exchange = mt5.symbol_info(convertto2).bid
                    else:
                        exchange = mt5.symbol_info(convertto).bid
                lot = round(riskusd/(stpips*exchange),2)
                request = {
                    "action": mt5.TRADE_ACTION_DEAL,
                    "symbol": symbol,
                    "volume": lot,
                    "type": mt5.ORDER_TYPE_SELL,
                    "price": price,
                    "sl": l,
                    "tp": k,
                    "deviation": deviation,
                    "magic": 234000,
                    "comment": "BRUTUS Mark V",
                    "type_time": mt5.ORDER_TIME_GTC,
                    "type_filling": mt5.ORDER_FILLING_IOC,
                }
            extraj = j - m
            selected=mt5.symbol_select(h,True)
            if not selected:
                print("Failed to select pair")
                mt5.shutdown()
                quit()
                iterator = 30
            lasttickb=mt5.symbol_info_tick(h).bid
            if lasttickb > extraj:
                RATE2.append(lasttickb)
            else:
                RATE2 = []
            if lasttickb > j:
                RATE.append(lasttickb)
            else:
                RATE = []
            print(RATE)
            lrate = len(RATE)
            if lrate >= threshold:
                result = mt5.order_send(request)
                print("1. order_send(): by {} {} lots at {} with deviation={} points".format(symbol,lot,price,deviation));
                iterator = 30
            lrate2 = len(RATE2)
            if lrate2 >= threshold2:
                result = mt5.order_send(request)
                print("1. order_send(): by {} {} lots at {} with deviation={} points".format(symbol,lot,price,deviation));
                iterator = 30

            if result.retcode != mt5.TRADE_RETCODE_DONE:
                print("2. order_send failed, retcode={}".format(result.retcode))
                result_dict=result._asdict()
                for field in result_dict.keys():
                    print("   {}={}".format(field,result_dict[field]))
                    if field=="request":
                        traderequest_dict=result_dict[field]._asdict()
                        for tradereq_filed in traderequest_dict:
                            print("       traderequest: {}={}".format(tradereq_filed,traderequest_dict[tradereq_filed]))
                print("shutdown() and quit")
                mt5.shutdown()
                quit()
                
        tend = dt.now()
        telapsed = tend - tstart
        telapsed = elapsed.total_seconds()
        telapsed = str(telapsed)
        ttimer = telapsed + '   seconds'
        TTIMER.append(ttimer)
        time.sleep(10)
    
    print(TTIMER)
    
```
